```python
import requests
newsurl="http://news.sina.com.cn/china/"
res=requests.get(newsurl)       #get方法
#print(res.encoding)    #读入时网页编码方式，乱码。可能与网页实际的编码不一致，需要进行修改
res.encoding='utf-8'  #修改读入时网页编码方式
#print(res.text)      #取出网页内容
#print(type(res))      #读入类型
```

```python
print(res.text) 
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(res.text,"html.parser")
```

```python
#print(soup.text)      #所有文字内容 
#print(type(soup))     #soup的类型
```

```python
news = soup.select('.news-item')[0]
news
```

```python
time = news.select('.time')
time
```

```python
#class里面的内容必须要在前加'.' 
#<>内的固定标签，那只需要在''内写标签内容，即可
for news in soup.select('.news-item'): #按列表进行(<div class="news-item first-news-item ">以news-item结尾)
    if(len(news.select('h2')) > 0):    #只要标题存在，就进行内容读取
        h2 = news.select('h2')[0].text
        time = news.select('.time')[0].text
        a = news.select('a')[0]['href']
        print(time,h2,a)
```

# 取出特定标签的元素

```python
#找出soup中含'h1'标签的元素
header=soup.select('h1')
print(header)
print(header[0].text)  #含'h1'标签的第一条语句[0]中的文本text
```

```python
#找出soup中含'a'标签的元素
alink=soup.select('a')
print(alink)
for link in alink
    print(link.text)  #每条语句中的文本text
```

# 取出含有特定CSS属性的元素

```python
# 使用select找出所有id为title的元素（id前面需要加#）id='title'
alink=soup.select('#title')
print(alink)
```

```python
# 使用select找出所有class为link的元素（class前面需要加.）class='link'
for link in soup.select('.link')
    print(link)  
```

# 取出所有a标签内的链接

```python
# 使用select找出所有a tag的href链接
alink=soup.select('a')
for link in soup.select('.link')
    print(link['href'])           #a标签里面的'href'的值，其他内容同理
```
